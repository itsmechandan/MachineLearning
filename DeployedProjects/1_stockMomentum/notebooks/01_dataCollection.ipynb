```python
import pandas as pd
import numpy as np
import yfinance as yf
from datetime import datetime 
from pathlib import Path
```

```python
TICKERS = [
    "TCS.NS",
    "INFY.NS",
    "WIPRO.NS",
    "HCLTECH.NS",
    "TECHM.NS",
    "LTIM.NS"
]

START_DATE = "2016-08-01"
END_DATE = datetime.today().strftime("%Y-%m-%d")

RAW_DATA_PATH = Path("/Users/venkatchandan/Desktop/ML_Projects/DeployedProjects/1_stockMomentum/data/raw/base_ohlcv_data.csv")
RAW_DATA_PATH.parent.mkdir(parents=True, exist_ok=True) 
```

```python

print("Downloading data for:", TICKERS)
data = yf.download(TICKERS, start=START_DATE, end=END_DATE)

print("Raw shape:", data.shape)
print("Raw columns:", data.columns)
```

```python
long_df = data.stack(level=1).reset_index()
long_df.columns
```

```python
long_df = long_df.rename(
    columns={
        "Date": "date",
        "Ticker": "symbol",
        "Open": "open",
        "High": "high",
        "Low": "low",
        "Close": "close",
        "Volume": "volume",
    }
)
base_df = long_df[["symbol", "date", "open", "high", "low", "close", "volume"]]

# 5. Clean types & sort
base_df["date"] = pd.to_datetime(base_df["date"])
base_df = base_df.sort_values(["symbol", "date"]).reset_index(drop=True)

print("Final shape:", base_df.shape)
base_df.tail()
```

```python
base_df.to_csv(RAW_DATA_PATH, index=False)
print("Saved to:", RAW_DATA_PATH)
```
